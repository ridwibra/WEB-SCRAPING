Scraping Reviews of Acer Aspire 5 Slim Lapto from Amazon Website

And saving the reviews in a csv file

```python
import re
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import numpy as np
```

Scraping positive reviews

```python
url = 'https://www.amazon.com/Acer-Display-Graphics-Keyboard-A515-43-R19L/product-reviews/B07RF1XD36/ref=cm_cr_arp_d_viewpnt_lft?ie=UTF8&filterByStar=positive&reviewerType=all_reviews'
```

```python
header = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36', 'Referer':'https://www.amazon.com/'}
```

```python
csv_file = open('positive_reviews.csv', 'w', encoding='utf-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Reviewer', 'Rating', 'Title','Body'])

for page in range(1, 100):
    html = requests.get(url + str(page), headers=header)
    soup = BeautifulSoup(html.content, 'lxml')
    for review in soup.find_all('div', class_='a-section review aok-relative'):
        reviewer = review.find('span', class_ = 'a-profile-name').text
        print(reviewer)
        rating =review.find('a', class_='a-link-normal').span.text.split()[0]
        print(rating)
        title= review.find('a', class_ ='a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold').span.text
        print(title)
        body = review.find('span', class_='a-size-base review-text review-text-content').span.text.split('\n')[1]
        print(body)
        csv_writer.writerow([reviewer, rating, title, body])

        print('\n\n')
```
